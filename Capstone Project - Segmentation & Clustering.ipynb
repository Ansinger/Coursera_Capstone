```python
#Importing Libraries
from geopy.geocoders import Nominatim
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
import folium 
```

```python
#Using GeoLocator to Determine Lat/Lon for Toronto
address = 'Toronto'
geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
```

```python
# Create Toroto Map Using Lat/Lon
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

#Add Markers to the Map based on Wiki Dataframe
for lat, lng, borough, neighborhood in zip(df_wiki['Latitude'], df_wiki['Longitude'], df_wiki['Boroughs'], df_wiki['Neighborhoods']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

```python
#Group by Neighborhoods & Reset the Index
toronto_grouped = df_wiki.groupby('Neighborhoods').mean().reset_index()
toronto_grouped
```

```python
#Define number of Clusters for Clustering the Data
kclusters = 4
toronto_grouped_clustering = toronto_grouped.drop('Neighborhoods', 1)

#Run K_means 
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

#check cluster
print(kmeans.labels_)
```

```python
toronto_grouped['Labels'] = kmeans.labels_
```

```python
toronto_grouped.groupby('Labels').mean()
```

```python
# Creating the Cluster Map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# Color Scheme for the Cluster Map
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_grouped['Latitude'], toronto_grouped['Longitude'], toronto_grouped['Neighborhoods'], toronto_grouped['Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
